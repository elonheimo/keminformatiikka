Hansch ja muut huomasivat, ett yhdisteen pienetkin rakenteelliset muutokset aiheuttivat kasvin auksiinin (kasvuhormoni) toiminnan ja vaikutuksen korostumista. He siis lysivt yhdisteit, jotka olivat **biologisesti aktiivisia** eli aiheuttivat biologisen vasteen. 

Tm aineen rakenteen suhde biologiaan oli matemaattisesti todettavissa korrelaationa, eli muuttujien keskinisen riippuvuutena. Hansch nimesi sen kvantitatiiviseksi rakenne-aktiivisuussuhteeksi (engl. _Quantitative Structure-activity-relationship_, QSAR).

### Yhdisteiden esittminen tietokoneille: SMILES

Oletko ikin miettinyt, miten tietokone ymmrt piirtmisi molekyylej? 

Yhdisteit, eli atomeja ja niiden vlisi sidoksia, on pyritty kuvaamaan matemaattisella kielell tietokoneelle erilaisilla menetelmill. Tm osoittautui erityisen haastavaksi matemaattiseksi mallinnustyksi, jossa erilaiset tylt ja puutteelliset tavat viesti tietokoneelle yhdisteist vasta useiden tyliden ja usein puutteellisten tietoteknisten yritysten jlkeen, 1980-luvulla kolmen veljeksen yritys Daylight informtics onnistui kehittmn tiedostotyypin nimelt: SMILES.

SMILES (Simplified Molecular Input Line Entry System) on ylivertaisen yksinkertainen tapa viesti molekyylirakenteesta yksinkertaisella, mutta kattavalla tavalla kirjoittaa, lukea ja tallentaa tietoteknisi kuvauksia kemiallisista yhdisteist. SMILES koostuu alkuaineiden ja numeroiden sarjasta, jota luetaan tietyll tavalla muodostaen, graafi, eli pisteiden ja janojen verkko. Tm verkko nojaa matemaattiseen verkkoteoriaan, ja se rakennetaan erilaisia algoritmeilla ja snttaulukolla.

#### PUG-REST menetelm

Ohjelmistoympristn voi tuoda helposti yhdisteit teettmll "kyselyit" PubChem-palvelimelta. Tm on erityisen hydyllist, kun halutaan nopeasti tuoda tietoa lkeyhdisteist ilman hakukoneita luettelemalla haluttuja molekyylej osaksi koodia.

```python
# Tee lista tuntemistasi lkemolekyyleist thn englannin kielell ja erota jokainen nimike pilkulla ja lainausmerkeill seuraavanlaisesti.

# Esimerkiksi, kirjoita englannin kielell.
compound_names = ["Aspirin","Paracetamol", "Venlafaxin", "Thalidomide", "Benzamide", "Cytosine"] 
```

```python
import requests
 # l muokkaa alla olevaa koodia. 
def retrieve_names(compound_names):
    pugrest = "https://pubchem.ncbi.nlm.nih.gov/rest/pug"
    pugoper = "property/CanonicalSMILES"
    pugout = "txt"
    results = {}
    for compound_name in compound_names:
        pugin = "compound/name/" + compound_name
        url = f"{pugrest}/{pugin}/{pugoper}/{pugout}"
        response = requests.get(url)
        if response.status_code == 200:
            results[compound_name] = response.text.strip()  # Strip to remove leading/trailing whitespaces
        else:
            results[compound_name] = f"Error: {response.status_code}"
    return results

# Muokkaa vain alla olevaa koodirivi.
compound_names = ["Aspirin","Paracetamol", "Venlafaxin", "Cocaine", "Benzamide", "Cytosine"] 
#Muokkaa vain yll olevaa koodirivi.
canonical_smiles = retrieve_names(compound_names)
for compound_name, smiles in canonical_smiles.items():
    print(f"Canonical SMILES for {compound_name}: {smiles}")
```

```python
# Voit mys muuttaa SMILES merkkijonon isomeeriseksi seuraavalla komentosarjalla. 
from rdkit import Chem
from rdkit.Chem import MolFromSmiles, MolToSmiles

def canonical_to_isomeric(smiles):
    mol = MolFromSmiles(smiles)
    if mol is not None:
        Chem.SanitizeMol(mol)
        isomeric_smiles = MolToSmiles(mol, isomericSmiles=True)
        return isomeric_smiles
    else:
        return None

# Muokkaa vain alla olevaa rivi
canonical_smiles = "CN1C2CCC1C(C(C2)OC(=O)C3=CC=CC=C3)C(=O)OC"
isomeric_smiles = canonical_to_isomeric(canonical_smiles)
print("Canonical SMILES:", canonical_smiles)
print("Isomeric SMILES:", isomeric_smiles)
```

Thn kysymykseen Lipinksin johtama tutkimusryhm etsi vastausta aloittaen suuren metatutkimuksen erilaisista patentoiduista lkkeist:

#### Lipinskin viiden snt

Christopher Lipinski tutki tiimins kanssa n. 2000 lkkeen dataa ja osoitti, ett niill on yleisesti ptevi hyvi piirteit: 

- Molekyylimassa on alle 500 Da, jossa 1 Da vastaa noin 1 atomimassayksikk 
- Jakautumiskerroin logP on alle 5
- Vetydonorien lukumr on alle 5, jossa vetydonori on atomi/rakenne, joka vetysidoksen luovuttuaan toimii negatiivisella osittaisvarauksella (Esim. happi, O vesimolekyylien vlisiss vetysidoksissa)
- Vetyakseptorien lukumr on alle 10, eli vetydonorien vastakohta siis pos. osittaisvarauksellinen atomi

#### Lipinski ei ole aina oikeassa

Nykyn tieteilijt mys kritisoivat Lipinskin viiden snt siit, ett se painottaa jokaista snt yht suuresti. Tt tukevcaa poikkeavaa ajattelua on olemassa myssen takia, ett jotkin yhdisteet toimivat loistavina lkkein, vaikka ne rikkovat Lipinskin viiden snt.

**Daptomysiini**: Daptomysiini on lipopeptidiantibiootti, jota kytetn vakavien bakteeri-infektioiden hoitoon. Sen molekyylipaino on suhteellisen suuri (yli 1000 Da) ja korkea LogP-arvo sen rasvaliukoisen luonteen vuoksi. Nist ominaisuuksista huolimatta daptomysiini pysyy tehokkaana, koska se vaikuttaa pasiassa infektion kohteena olevassa kohdassa, miss se pystyy tunkeutumaan bakteerisolukalvoihin ja vaikuttamaan antibakteerisesti.

**Vancomysiini**: Vancomysiini on toinen antibiootti, joka ylitt Lipinskin molekyylipainon suosituksen. Se on glykoproteiiniantibiootti, jota kytetn Gram-positiivisten bakteerien aiheuttamien infektioiden hoitoon. Vancomysiini toimii estmll soluseinn synteesi bakteereissa. Sen suuri koko ja hydrofiilinen luonne eivt est sen aktiivisuutta kohdebakteereita vastaan, kun sit annetaan asianmukaisesti.

**Siklosporiini**: Siklosporiini on immunosuppressiivinen lke, jota kytetn hylkimisen estoon elinsiirroissa sek autoimmuunisairauksien, kuten nivelreuman ja psoriasiksen, hoitoon. Siin on suuri molekyylipaino ja LogP-arvo, mutta se on muotoiltu tavalla, joka mahdollistaa tehokkaan toimituksen kohdekudoksiin, kuten immuunivasteeseen osallistuviin lymfosyytteihin.

**Takrolimuusi**: Samoin kuin siklosporiini, takrolimuusi on immunosuppressiivinen lke, jota kytetn elinsiirroissa ja autoimmuunisairauksissa. Siin on ominaisuuksia, jotka poikkeavat Lipinskin snnist, mutta se on muotoiltu saavuttamaan terapeuttiset pitoisuudet immuunisoluissa samalla kun se minimoi systeemiset sivuvaikutukset.

### Regressioanalyysi

Lineeariset sovitukset ovat tuttuja varmasti matematiikan tunneilta. Tm matemaattinen mallintaminen on yksi keskeisimpi menetelmi keminformatiikassa koska sill voidaan tarkastella erilaisten muuttjien suhdetta toisiinsa (korrelaatio) aloittaen alkaen kahdesta tai usemmasta tekijst. Alla on koodia, jolla voidaan sovittaa regressiomalli alkaanien ja sykloalkaaninien datasettiin. Mit huomaat kun koodi tulostaa graafin? Onko tm tarkka malli?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Lataa dataa .csv tiedostosta taulukoksi
df = pd.read_csv('BP.csv', encoding='latin-1')

# Sovita lineaarinen regressiomalli 
model = LinearRegression()
X = df[['Moolimassa']]  # Feature (independent variable)
y = df['Kiehumispiste_K']  # Target variable (dependent variable)
model.fit(X, y)

# Piirr  pistekuvio datasta 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Moolimassa', y='Kiehumispiste_K', data=df, color='blue', label='Data Points')

# Piirr regressiosuora
plt.plot(X, model.predict(X), color='red', label='Linear Regression')

# Aseta otsikot ja mitat
plt.xlabel('Moolimassa')
plt.ylabel('Kiehumispiste (K)')
plt.title('Linear Regression Model')

# Show legend
plt.legend()

# Nyt sovitus
plt.show()
```
